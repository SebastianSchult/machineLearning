```python
# ðŸ“Š Exploratory Data Analysis: adult_cleaned.csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Notebook-Einstellungen
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Daten laden
df = pd.read_csv("adult/adult_cleaned.csv")
```

```python
# Zielspalte (zur Erinnerung)
print("ðŸ’° Zielverteilung:")
print(df["income"].value_counts(normalize=True))
```

```python
# âž¤ Summary-Statistik (numerisch)
print("ðŸ“ˆ Numerische Ãœbersicht:")
display(df.describe())
```

```python

# âž¤ Einkommensverteilung
plt.figure(figsize=(5,4))
sns.countplot(data=df, x="income")
plt.title("Verteilung Einkommen (0 = <=50K, 1 = >50K)")
plt.show()
```

```python
# âž¤ Einkommen nach Bildungsgrad
plt.figure(figsize=(12,5))
sns.barplot(data=df, x="education", y="income", estimator=lambda x: sum(x)/len(x))
plt.xticks(rotation=45)
plt.title("â« Durchschnittliches Einkommen je Bildungsgrad")
plt.ylabel("Anteil >50K")
plt.show()
```

```python
# âž¤ Einkommen nach Geschlecht
plt.figure(figsize=(5,4))
sns.barplot(data=df, x="sex", y="income", estimator=lambda x: sum(x)/len(x))
plt.title("Einkommen nach Geschlecht (Anteil >50K)")
plt.ylabel("Anteil >50K")
plt.show()
```

```python
# âž¤ Alter vs Einkommen (Boxplot)
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="income", y="age")
plt.title("Alter vs Einkommen")
plt.xlabel("Income (0 = <=50K, 1 = >50K)")
plt.show()
```

```python
# âž¤ Korrelationen numerischer Merkmale
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("ðŸ“Š Korrelationsmatrix")
plt.show()
```
